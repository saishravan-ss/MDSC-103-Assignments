```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x1 = np.linspace(-5,5,100)
x2 = np.linspace(-5,5,100)
fun = lambda x1,x2: 4*x1 + 6*x2 - 2*(x1)**2 - 2*(x1)*(x2) - 2*(x2)**2
plt.plot(x1,x2,fun(x1,x2))
plt.show()
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))                                       # setting the figure size for the plot
ax = fig.add_subplot(1,2,1,projection='3d')                                        # adding subplot with the projection 3d
x1 = np.linspace(-5,10,100)                                                        # giving the limits for the x1 from -5 to 10
x2 = np.linspace(-5,10,100)                                                        # giving the limits for the x1 from -5 to 10
x1,x2 = np.meshgrid(x1,x2)                                                         # forming the meshgrid with x1 and x2
fun = lambda x1,x2: 4*x1 + 6*x2 - 2*(x1)**2 - 2*(x1)*(x2) - 2*(x2)**2              # defining the function using lambda
ax.scatter3D(x1,x2,fun(x1,x2),marker='*',color='pink')                             # plotting the 3d-plot using scattered
plt.xlabel("X1")                                                                   # setting x-label as X1
plt.ylabel("X2")                                                                   # setting x-label as X2
plt.title("Plot for the 4*x1 + 6*x2 - 2*(x1)**2 - 2*(x1)*(x2) - 2*(x2)**2")        # setting the title for the plot
plt.show()
```
