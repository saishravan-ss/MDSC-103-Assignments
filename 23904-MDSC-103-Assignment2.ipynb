Given to plot the graphs of the x^2 and x^3 convergence functions using Gradient Descent method

```python
# First plotting the graph for the y = x^2 function

import numpy as np
import matplotlib.pyplot as plt

f = lambda x:x**2
z = np.linspace(-5,5,100)
plt.plot(z,f(z))
plt.title('Plot of y = x^2 function')
plt.xlabel('x values from -5 to +5')
plt.ylabel('x^2 values')
plt.show()
```

```python
f = lambda x:x**2
z = np.linspace(-5,5,100)
plt.plot(z,f(z))

x0= 4 # setting initial point to 4
l = 0.1
d = 1e-5
GD = lambda x0:x0-(0.1)* 2 * x0  # defining the gradient method formula and we apply this for the 100 iterations since d = 1e-5
for i in range(100):
    x1 = GD(x0)
    if((x0-x1)<d):
        print(x1)   # prints the value of x1 where the convergence takes place
        break
    plt.scatter(x1,f(x1),color='r')
    plt.title('Convergence of x^2 without using function')
    plt.xlabel('x values from -5 to +5')
    plt.ylabel('x^2 values')
    x0 = x1
plt.show()
```

Now writing the same above method in the form of function where user will give the function,first derivative function,initial point,
learning rate value and delta value as the input.

Then we plot the graph of the given function w.r.t x values and then apply the gradient descent method where we again plot the scatter
of the x1 and f(x1) values on the same above plotted plot.

```python
def GD_fun(fun,df,ip,lr,tol):   
    x = np.linspace(-5,5,100)
    plt.plot(x,fun(x))
    plt.grid()
#     plt.title('Convergence of x^2 functions using Gradient Descent method')
    plt.xlabel('-5 to +5 values ')
    plt.ylabel('Squares of x values')
    GD = lambda x0,lr:x0-(lr)*df(x0)
    for i in range(100):
        x1 = GD(ip,lr)
        plt.scatter(x1,fun(x1),color='r')
        plt.title('Convergence of x^2 and x^3 functions using Gradient Descent method')

        if(np.abs(ip-x1)<tol):
            print(x1)
            break
        ip = x1
    plt.show()
```

Convergence of the function 
    $ fun = lambda x: x^2 $

```python
fun = lambda x: x**2
df = lambda x : 2*x
GD_fun(fun,df,4,0.1,1e-5)
```

Convergence of the function
    $ fun = lambda x: x^3 $

```python
fun = lambda x: x**3
df = lambda x: 3*x**2
GD_fun(fun,df,2,0.1,1e-5)
```
